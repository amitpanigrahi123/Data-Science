## PIVOT TABLE

```python
df1 = pd.read_csv('weather2.csv')
df1
```

## here we can see that there are two temperature records in one date, for each cities...................so  lets create a dataframe which consists an average temperature throughout the day, for all the dates

```python
df1.pivot_table(index ='city',columns ='date')   #use of pivot_table funct.
```

```python
df1.pivot_table(index='city', columns='date',aggfunc =sum)  #aggregate addition of the values
```

```python
df1.pivot_table(index='city', columns='date',aggfunc ='count')  #it shows 2 values(each category) in each dates for each city
```

```python
df1.pivot_table(index='city', columns='date',margins = True)   #it has taken the mean of all the dates for each cities and all
```

## Grouper

```python
df2 = pd.read_csv('weather3.csv')
df2
```

## suppose we want to know the average temperature for the month of may and december for newyork city, so we can use grouper function to do it

## before grouping we need to convert the date column  to date format, coz bydefault its in string format

```python
df2['date']=pd.to_datetime(df2['date'])
df2
```

## so now its in date format

```python
df2.pivot_table(index=pd.Grouper(freq='M',key ='date'),columns='city')   #on date we applied the 'Grouper' function
                                                                          #columns as city
                                                                         #freq is the keyword by which we group and here we took
                                                                         # 'M' because it specifies frequency by month
```

## so here we got the average temp. and humidity of the may month and december month of new york city
